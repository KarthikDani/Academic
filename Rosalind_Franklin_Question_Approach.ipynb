```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

confidence_level = 0.95

t = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Plot the CDF of the standard normal distribution
x = np.linspace(-5, 5, 1000)
cdf = stats.norm.cdf(x)

plt.plot(x, cdf, label='CDF of Standard Normal Distribution')
plt.axvline(x=t, color='red', linestyle='--', label=f'Critical Value (t) = {t:.2f}')
plt.legend()
plt.title('Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid()
plt.show()

print(f'The critical value (t) for a {confidence_level*100:.2f}% confidence interval is approximately {t:.2f}.')
```
